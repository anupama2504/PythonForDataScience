# Predicting humidity at 3 pm using Classification Tree 

```python
clean_data = dat.copy()
clean_data['relative_humidity_3pm'].describe()
#considering 50% and above relative humidity values to be high humidity
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm'] > 24.37)*1
# clean_data['high_humidity_label']
y = clean_data[['high_humidity_label']].copy()
y.head()
```

```python
cols = clean_data.columns
# print(cols)
cols.shape
cols[:9]
```

```python
#selecting 9 am sensor signals as features to predict Humidity at 3 pm
x = clean_data[cols[:9]]
x.columns
```

# Perform Test and Train Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 324)
```

# Fit on Train set

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(x_train, y_train)
```

# Predict on Test Set

```python
predictions = humidity_classifier.predict(x_test)
```

# Measure accuracy

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```
